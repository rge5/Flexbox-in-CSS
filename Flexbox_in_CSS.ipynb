```python
1. Describe the main differences between the CSS Flexbox layout model and the CSS Grid layout model. When
would you choose to use one over the other?
->
| Feature                | **Flexbox**                                           | **Grid**                                                                  |
| ---------------------- | ----------------------------------------------------- | ------------------------------------------------------------------------- |
| **Direction**          | One-dimensional layout (either **row** or **column**) | Two-dimensional layout (**rows and columns**)                             |
| **Content Control**    | Aligns items along a **single axis**                  | Controls both **rows and columns** simultaneously                         |
| **Layout Focus**       | Best for **aligning items in a line**                 | Best for **overall page layout or complex grids**                         |
| **Item Placement**     | Items placed in **source order**                      | Items can be **placed anywhere in the grid**, independent of source order |
| **Ease of Use**        | Easier for **small layouts or navbars**               | Better for **large, structured layouts**                                  |
| **Alignment Features** | Great for **alignment, spacing, and distribution**    | More powerful for **precise placement**                                   |

```

2.  Explain the role of the following key properties in the Flexbox layout mode.

->  
1.   justify-content -Aligns flex items along the main axis (horizontal by default).
2.   align-item- Align flex item along th cross axis (verticle by default).
3.   gap- space between  flex items.
4.   flex-direction- define the main axis direction --the direction in which flex items are laid out.
5. flex-wrap- control wether flex items should wrap to a new line.
